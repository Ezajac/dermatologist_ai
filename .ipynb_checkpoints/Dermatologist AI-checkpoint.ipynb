```python
def add_new_last_layer(base_model, nb_classes, fc_size, fc_size2):
    """Add last layer to the convnet
    Args:
    base_model: keras model excluding top
    nb_classes: # of classes
    Returns:
    new keras model with last layer
    """
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(fc_size, activation='relu')(x) #new FC layer, random init
    x = Dropout(0.4)(x)
    x = Dense(fc_size2, activation='relu')(x) #new FC layer, random init
    predictions = Dense(nb_classes, activation='softmax')(x) #new softmax layer
    model = Model(input=base_model.input, output=predictions)
    return model
```

```python
nb_classes = 3
fc_size = 1024
fc_size2 = 512
base_model = InceptionV3(weights='imagenet', include_top=False) #include_top=False excludes final FC layer
model = add_new_last_layer(base_model, nb_classes, fc_size, fc_size2)
```

```python
model.save_weights('saved_models/inception_baseline.h5')
```

```python
def setup_to_transfer_learn(model, base_model):
    """Freeze all layers and compile the model"""
    for layer in model.layers[:311]:
        layer.trainable = False
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
setup_to_transfer_learn(model, base_model)
#Other Optimizers 'adam'. Not much success with SGD(lr=0.0001, momentum=0.9), 'rmsprop'
```

```python
epochs = 20
batch_size = batch_size

### Do NOT modify the code below this line.

checkpointer = ModelCheckpoint(filepath='saved_models/weights.best.from_scratch.hdf5', 
                               verbose=1, save_best_only=True)
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0.01, patience=3, verbose=1, mode='auto')

hist = model.fit(np.array(trn_files), np.array(trn_label_vecs), 
          validation_data=(np.array(tst_files), np.array(tst_label_vecs)),
          epochs=epochs, batch_size=batch_size, callbacks=[checkpointer, early_stopping], verbose=1)
```
